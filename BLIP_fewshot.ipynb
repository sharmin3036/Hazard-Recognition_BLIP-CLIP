```python
# ===============================
# 🔄 Cell 2: Setup & Model Loading
# ===============================
from transformers import Blip2Processor, Blip2ForConditionalGeneration
from PIL import Image
import torch
import os
from google.colab import drive
import pandas as pd
from datetime import datetime
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import logging
logging.getLogger("transformers").setLevel(logging.ERROR)
```

```python
# Paths
base_path = '/content/drive/MyDrive/TRB2026_sharmin/css-data'
subfolders = ['test', 'valid']
```

```python
# Load BLIP-2 model
# model_version = "Salesforce/blip2-opt-2.7b"
model_version = "Salesforce/blip2-flan-t5-xl" # Use a smaller model to avoid OOM
try:
    processor = Blip2Processor.from_pretrained(model_version)
    model = Blip2ForConditionalGeneration.from_pretrained(
        model_version,
        torch_dtype=torch.float16,
        device_map="auto"  # Better GPU memory management
    )
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"✅ Model loaded successfully on {device}")
except Exception as e:
    print(f"❌ Error loading model: {e}")
    raise
```

```python
# Declare result folder name (change per run if needed)
result_folder_name = "result_blip_t5_fewshot"
```

```python
# ===============================
# 🏷️ Prompts & Few-shot Examples (kept as you provided)
# ===============================
label_few_shots = {
    0: [
        ("Is the worker wearing a helmet?", "Yes"),
        ("Is there anyone not wearing a helmet?", "No"),
        ("Is a helmet visible?", "Yes"),
        ("Is someone missing head protection?", "No")
    ],
    7: [
        ("Is the worker wearing a safety vest?", "Yes"),
        ("Is anyone missing a vest?", "No"),
        ("Is a reflective vest visible?", "Yes"),
        ("Is the person not wearing safety gear?", "No")
    ],
    6: [
        ("Is there a traffic cone in the image?", "Yes"),
        ("Can you see a safety cone?", "Yes"),
        ("Is a cone present?", "Yes"),
        ("Is there a warning cone?", "Yes")
    ],
    8: [
        ("Is construction machinery present?", "Yes"),
        ("Do you see any machines?", "Yes"),
        ("Is there heavy equipment in the scene?", "Yes"),
        ("Is construction equipment visible?", "Yes")
    ],
    9: [
        ("Is a construction vehicle in the image?", "Yes"),
        ("Can you spot a construction truck?", "Yes"),
        ("Is there a work vehicle?", "Yes"),
        ("Is heavy vehicle visible?", "Yes")
    ]
}

label_prompts = {
    0: "Is there a person wearing a helmet in this image?",
    7: "Is someone wearing a safety vest in the image?",
    6: "Can you see a traffic cone in the image?",
    8: "Is there construction equipment in this image?",
    9: "Is a construction vehicle visible in the image?"
}

label_names = {
    0: "Helmet",
    7: "Safety Vest",
    6: "Safety Cone",
    8: "Machinery",
    9: "Vehicle"
}
```

```python
# ===============================
# 🧠 Helper: Few-shot Caption→QA (minimal patch)
# ===============================
def ask_question(image_path, question, label_id=None):
    image = Image.open(image_path).convert("RGB")

    # 1) Generate image caption
    caption_prompt = "Describe the image in detail."
    cap_inputs = processor(images=image, text=caption_prompt, return_tensors="pt")
    cap_inputs = {k: v.to(model.device) for k, v in cap_inputs.items()}  # move to device only
    cap_output = model.generate(**cap_inputs)
    caption = processor.decode(cap_output[0], skip_special_tokens=True).strip()

    # 2) Insert label-specific few-shot examples
    few_shot_qas = ""
    if label_id in label_few_shots:
        for q, a in label_few_shots[label_id]:
            few_shot_qas += f"Question: {q} Answer: {a}\n"

    # 3) Final prompt (caption → few-shot → question)
    full_prompt = f"Image description: {caption}\n{few_shot_qas}Question: {question} Answer:"
    qa_inputs = processor(images=image, text=full_prompt, return_tensors="pt")
    qa_inputs = {k: v.to(model.device) for k, v in qa_inputs.items()}  # move to device only
    qa_output = model.generate(**qa_inputs)
    answer = processor.decode(qa_output[0], skip_special_tokens=True).strip()

    return "yes" if answer.lower().startswith("yes") else "no"
```

```python
# ===============================
# 🧪 Cell 3: Inference & Label Saving (Few-shot)
# ===============================
total_images = 0
processed_images = 0

for sub in subfolders:
    image_folder = os.path.join(base_path, sub, "images")
    output_label_folder = os.path.join(base_path, sub, result_folder_name)

    # Check if folders exist
    if not os.path.exists(image_folder):
        print(f"⚠️  Warning: {image_folder} does not exist. Skipping...")
        continue

    os.makedirs(output_label_folder, exist_ok=True)

    # Get list of image files
    image_files = [f for f in os.listdir(image_folder)
                   if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff'))]

    print(f"📁 Processing {len(image_files)} images in {sub} folder...")

    for img_file in tqdm(image_files, desc=f"Processing {sub}"):
        total_images += 1
        img_path = os.path.join(image_folder, img_file)

        # Check if file exists and is readable
        if not os.path.exists(img_path):
            print(f"⚠️  Warning: {img_path} does not exist. Skipping...")
            continue

        predicted_labels = []

        for label_id, prompt in label_prompts.items():
            # Few-shot: pass label_id so the function can include label-specific Q/A examples
            response = ask_question(img_path, prompt, label_id=label_id)
            if response == "yes":
                predicted_labels.append(label_id)

        predicted_labels = sorted(set(predicted_labels))
        label_filename = os.path.splitext(img_file)[0] + ".txt"
        output_path = os.path.join(output_label_folder, label_filename)

        try:
            with open(output_path, 'w') as f:
                for label in predicted_labels:
                    f.write(f"{label}\n")
            processed_images += 1
        except Exception as e:
            print(f"❌ Error writing labels for {img_file}: {e}")

print(f"✅ BLIP-2 predictions saved to '{result_folder_name}' folders.")
print(f"📊 Processed {processed_images}/{total_images} images successfully.")
```

```python
# Create export directory with timestamp
now = datetime.now().strftime("%Y%m%d_%H%M%S")
export_dir = os.path.join(base_path, f"{result_folder_name}_" + now)
os.makedirs(export_dir, exist_ok=True)
```

```python
# ===============================
# 📊 Cell 4: Evaluation & Visualization (Few-shot)
# ===============================
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import classification_report

true_labels, pred_labels = [], []
skipped_files = []

for sub in subfolders:
    image_dir = os.path.join(base_path, sub, "images")
    gt_dir = os.path.join(base_path, sub, "labels_superclean_GT")  # ✅ correct GT path
    pred_dir = os.path.join(base_path, sub, result_folder_name)

    # Check if directories exist
    if not all(os.path.exists(d) for d in [image_dir, gt_dir, pred_dir]):
        print(f"⚠️  Warning: Missing directories for {sub}. Skipping evaluation...")
        continue

    image_files = [f for f in os.listdir(image_dir)
                   if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff'))]

    for file in image_files:
        name = os.path.splitext(file)[0]
        gt_path = os.path.join(gt_dir, name + ".txt")
        pr_path = os.path.join(pred_dir, name + ".txt")

        if not os.path.exists(gt_path) or not os.path.exists(pr_path):
            skipped_files.append(file)
            continue

        try:
            # Read ground truth labels
            with open(gt_path) as f:
                gt = [int(x.strip()) for x in f if x.strip().isdigit()]

            # Read predicted labels
            with open(pr_path) as f:
                pr = [int(x.strip()) for x in f if x.strip().isdigit()]

            true_labels.append(gt)
            pred_labels.append(pr)

        except Exception as e:
            print(f"❌ Error reading labels for {file}: {e}")
            skipped_files.append(file)

print(f"📊 Evaluation dataset: {len(true_labels)} samples")
if skipped_files:
    print(f"⚠️  Skipped {len(skipped_files)} files due to missing labels")
```

```python
# ---- Few-shot friendly evaluation (no hard exit) ----
has_data = len(true_labels) > 0
if not has_data:
    print("❌ No valid data found for evaluation. Check your file paths and label formats.")
    # Create empty placeholders so logging can still proceed
    mlb = MultiLabelBinarizer(classes=sorted(label_prompts.keys()))
    mlb.fit([[]])  # just to set classes_
    y_true = np.zeros((0, len(mlb.classes_)), dtype=int)
    y_pred = np.zeros_like(y_true)
    micro_precision = micro_recall = micro_f1 = 0.0
    macro_precision = macro_recall = macro_f1 = 0.0
    # Minimal empty report frame
    report_df = pd.DataFrame(
        {k: [0.0, 0.0, 0.0, 0] for k in ["precision", "recall", "f1-score", "support"]},
        index=[label_names[l] for l in mlb.classes_]
    )
else:
    # Initialize MultiLabelBinarizer with all possible labels
    mlb = MultiLabelBinarizer(classes=sorted(label_prompts.keys()))
    y_true = mlb.fit_transform(true_labels)
    y_pred = mlb.transform(pred_labels)

    # Per-label Micro/Macro Evaluation
    micro_precision = precision_score(y_true, y_pred, average='micro', zero_division=0)
    micro_recall    = recall_score(y_true, y_pred, average='micro', zero_division=0)
    micro_f1        = f1_score(y_true, y_pred, average='micro', zero_division=0)

    macro_precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
    macro_recall    = recall_score(y_true, y_pred, average='macro', zero_division=0)
    macro_f1        = f1_score(y_true, y_pred, average='macro', zero_division=0)

    # Classification report
    report = classification_report(
        y_true, y_pred,
        target_names=[label_names[l] for l in mlb.classes_],
        output_dict=True,
        zero_division=0
    )
    report_df = pd.DataFrame(report).T.round(3)



# Save report
report_df_path = os.path.join(export_dir, "classification_report.csv")
report_df.to_csv(report_df_path)

# Print evaluation summary
print("\n✅ Evaluation Summary")
print("="*50)
print(report_df)
print(f"\nMicro F1: {micro_f1:.3f}, Precision: {micro_precision:.3f}, Recall: {micro_recall:.3f}")
print(f"Macro F1: {macro_f1:.3f}, Precision: {macro_precision:.3f}, Recall: {macro_recall:.3f}")

# Save summary CSV
summary_df = pd.DataFrame({
    'Metric': ['Micro Precision', 'Micro Recall', 'Micro F1', 'Macro Precision', 'Macro Recall', 'Macro F1'],
    'Score': [micro_precision, micro_recall, micro_f1, macro_precision, macro_recall, macro_f1]
}).round(3)
summary_df.to_csv(os.path.join(export_dir, "summary_metrics.csv"), index=False)
```

```python

# Per-label confusion matrices (only if we have data)
if has_data:
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()

    for i, label_id in enumerate(mlb.classes_):
        cm = confusion_matrix(y_true[:, i], y_pred[:, i], labels=[0, 1])
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i],
                    xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
        axes[i].set_title(f'{label_names[label_id]} (ID: {label_id})')
        axes[i].set_xlabel('Predicted')
        axes[i].set_ylabel('Actual')

        # Save each matrix separately
        fig_individual, ax_ind = plt.subplots(figsize=(6, 5))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax_ind,
                    xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
        ax_ind.set_title(f'{label_names[label_id]} Confusion Matrix')
        ax_ind.set_xlabel('Predicted')
        ax_ind.set_ylabel('Actual')
        fig_individual.tight_layout()
        fig_individual.savefig(os.path.join(export_dir, f"conf_matrix_label_{label_id}.png"), dpi=300)
        plt.close(fig_individual)

    # Hide any unused subplot axes
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.savefig(os.path.join(export_dir, "confusion_matrices_grid.png"), dpi=300, bbox_inches='tight')
    plt.show()
else:
    print("ℹ️ Skipping confusion matrices because no evaluation samples were available.")

# Persist the prompts and few-shot exemplars for reproducibility
import json as _json
with open(os.path.join(export_dir, "label_prompts.json"), "w") as _f:
    _json.dump(label_prompts, _f, indent=2)
try:
    with open(os.path.join(export_dir, "label_few_shots.json"), "w") as _f:
        _json.dump(label_few_shots, _f, indent=2)
except NameError:
    pass  # if label_few_shots not defined, ignore
```

```python
# ===============================
# 📝 Cell 5: Log Run Info to Summary Index (Few-shot)
# ===============================
import csv
import json

# Summary log file path
summary_log_path = os.path.join(base_path, "experiment_log.csv")

# Define run metadata
experiment_tag = result_folder_name
# Store label prompts; if you want to include few-shot exemplars in the log, add to 'notes'
label_prompts_str = json.dumps(label_prompts, indent=2)

# Derive a brief few-shot note (how many examples per label if available)
try:
    shots_per_label = {k: len(v) for k, v in label_few_shots.items()}
    avg_shots = round(sum(shots_per_label.values()) / max(len(shots_per_label), 1), 2)
    fewshot_note = f"Few-shot (avg {avg_shots} shots/label); caption→QA; strict yes/no."
except NameError:
    fewshot_note = "Few-shot (caption→QA; strict yes/no)."

notes = fewshot_note

summary_row = {
    "timestamp": now,
    "experiment": experiment_tag,
    "macro_precision": round(macro_precision, 3) if has_data else 0.0,
    "macro_recall": round(macro_recall, 3) if has_data else 0.0,
    "macro_f1": round(macro_f1, 3) if has_data else 0.0,
    "micro_precision": round(micro_precision, 3) if has_data else 0.0,
    "micro_recall": round(micro_recall, 3) if has_data else 0.0,
    "micro_f1": round(micro_f1, 3) if has_data else 0.0,
    "result_folder": export_dir,
    "model_version": model_version,
    "notes": notes,
    "prompt": label_prompts_str
}

# Write to CSV (append or create)
write_header = not os.path.exists(summary_log_path)
with open(summary_log_path, mode='a', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=list(summary_row.keys()))
    if write_header:
        writer.writeheader()
    writer.writerow(summary_row)

print(f"✅ Experiment log saved to {summary_log_path}")
```
